```python
df_fact_telemetry = spark.table("aerospace.bronze.brz_fact_telemetry")

# Display schema
df_fact_telemetry.printSchema()
```

```python
# Show few records
display(df_fact_telemetry.limit(25))
```

```python
# Summary statistics for numeric columns
display(df_fact_telemetry.describe())
```

```python
# Count of rows
row_count = df_fact_telemetry.count()
print(f"Row count: {row_count}")
```

```python
# Count of nulls per column
from pyspark.sql.functions import col, sum as spark_sum

null_counts = df_fact_telemetry.select([spark_sum(col(c).isNull().cast("int")).alias(c) for c in df_fact_telemetry.columns])
display(null_counts)
```

```python
# Distinct values for categorical columns
display(df_fact_telemetry.select("engine_id").distinct())
display(df_fact_telemetry.select("cycle").distinct())
```
